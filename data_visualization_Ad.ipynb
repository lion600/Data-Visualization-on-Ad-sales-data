```python
click = df['Clicks']
price = df['Search Keyword']

fig, ax = plt.subplots(figsize =(16, 9))

ax.barh(price,click)

for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)

ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')

ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)

ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)

ax.invert_yaxis()

for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
            str(round((i.get_width()), 2)),
            fontsize = 10, fontweight ='bold',
            color ='grey')

# Add Plot Title
ax.set_title('keyword search and number of clicks',
        loc ='left', )

# Add Text watermark
fig.text(0.9, 0.15, 'Jeeteshgavande30', fontsize = 12,
        color ='grey', ha ='right', va ='bottom',
        alpha = 0.7)

plt.show()
```

```python
#only 3 categories in keyword type,lets see which one is used most.
category_counts = df['Keyword type'].value_counts()

result_df = pd.DataFrame(category_counts).reset_index()
result_df.columns = ['Category', 'Count']

print(result_df)
```

```python
fig = plt.figure(figsize =(10, 7))
plt.pie(result_df['Count'], labels = result_df['Category'],shadow=True,
        startangle=90, autopct='%1.1f%%',
        wedgeprops={'edgecolor': 'black'})
plt.title('Keyword type') 
plt.show()
```

```python
# Set the plot style (optional)
sns.set(style="whitegrid")

# Create the faceted bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x='Campaign', y='Impressions', hue='Keyword type', data=df, ci=None)

# Add labels and title
plt.xlabel('Campaign')
plt.ylabel('Impression')
plt.title('Impression by Campaign and Keyword')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

All though Board keyword type is the popular one but it has heighest number of impression 

```python
# Group by 'Cost' and count the occurrences
grouped = df.groupby('Cost ($)').count()

# Sort the groups by 'Cost' in descending order and select the top 3
top_groups = grouped.sort_index(ascending=False).head(3)

# Create the bar graph for the top ad groups
plt.figure(figsize=(8, 6))
plt.bar(top_groups.index, top_groups['Ad group'], align='center', alpha=0.7)

# Add labels and title
plt.xlabel('Cost')
plt.ylabel('Number of Ad Groups')
plt.title('Top Ad Group(s) with Most Cost')

# Show the actual group names as X-axis labels
plt.xticks(top_groups.index, top_groups.index)

# Show the plot
plt.grid(axis='y')
plt.show()
```

```python
clean_df = df.dropna()
```

```python
clean_df.isnull().sum()
```

```python
clean_df
```

```python
clean_df = pd.DataFrame(clean_df)
clean_df['Cost ($)'] = clean_df['Cost ($)'].astype(float)
clean_df['Payment ($)'] = clean_df['Payment ($)'].replace({'\$': '', ',': ''}, regex=True).astype(float)


# Calculate the profit and create a new column 'Profit'
clean_df['Profit'] = clean_df['Payment ($)'] - clean_df['Cost ($)']

clean_df
```

```python
plt.figure(figsize=(10, 6))
plt.bar(clean_df['Ad group'],clean_df['Profit'], alpha=0.7)

plt.xlabel('Ad Group')
plt.ylabel('Profit')
plt.title('Profit per Ad Group')

plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()






```

```python
plt.figure(figsize=(10, 6))
plt.bar(clean_df['Campaign'],clean_df['Profit'], alpha=0.7)

plt.xlabel('Campaign')
plt.ylabel('Profit')
plt.title('Profit per Campaign')

plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()

```

```python
plt.figure(figsize=(10, 6))
plt.bar(clean_df['Keyword type'],clean_df['Profit'], alpha=0.7)

plt.xlabel('keyword type')
plt.ylabel('Profit')
plt.title('Profit per keyword type')

plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()

```

We can conclude that Ad group 'Reservation_Management_System_Phrase' is giving the highest profit which comes under EK_Generic_Tri-Tok_Reservation campaign.
